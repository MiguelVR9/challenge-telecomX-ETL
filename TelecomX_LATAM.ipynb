### Recuento de evasión por variables categóricas

```python
variables_categoricas = [
    "Género",
    "Tipo contrato",
    "Forma de pago",
    "Servicio de internet",
    "Tiene pareja",
    "Dependientes",
    "Factura digital",
    "Lineas múltiples",
    "Seguridad online",
    "Respaldo online",
    "Protección dispositivo",
    "Soporte técnico",
    "Televisión por cable",
    "Streaming películas"
]
```

> Se evaluarán las diferentes variables para ver su influencia con la variable de interés que son los que "Sí" cancelaron

```python
influencia_var_categoricas = []

for var in variables_categoricas:
    tabla = pd.crosstab(df[var], df["Evasión"], normalize="index").round(4) * 100
    tabla = tabla.rename(columns={0:"No", 1:"Sí"})

    # Diferencia máxima entre categorías de la variable
    diferencia_max = tabla["Sí"].max() - tabla["Sí"].min()

    influencia_var_categoricas.append({
        "Variable": var,
        "Evasión mínima(%)": tabla["Sí"].min(),
        "Evasión máxima(%)": tabla["Sí"].max(),
        "Diferencia máxima (%)": diferencia_max
    })

influencia_var_categoricas_df = pd.DataFrame(influencia_var_categoricas).sort_values(by="Diferencia máxima (%)", ascending=False)
influencia_var_categoricas_df
```

> A partir de los resultados obtenidos, se decidió considerar como variables relevantes para el estudio aquellas que presenten una diferencia mayor a **30%** entre el porcentaje correspondiente al valor mínimo de evasión y el porcentaje correspondiente al valor máximo de evasión.

## 1. Introducción

El objetivo principal de este análisis es identificar los factores críticos que impulsan la evasión de clientes (Churn) en Telecom X. La pérdida de suscriptores no solo representa una disminución en los ingresos, sino que también incrementa los costos operativos debido a la inversión necesaria para adquirir nuevos usuarios. Mediante este estudio, se busca transformar datos operativos en información estratégica que permita generar insights y sugerencias para ayudar a Telecom X a retener más clientes .
