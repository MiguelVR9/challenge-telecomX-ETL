##📌 Extracción

```python
#Cargar los datos
url_json = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
```

```python
# Convertir los datos a un DataFrame de Pandas
df = pd.read_json(url_json)
```

### Comprobación de inconsistencias en los datos

```python
#Tratmiento de los datos anidados (normalizándolos)
customer_df = pd.json_normalize(df["customer"])
phone_df = pd.json_normalize(df["phone"])
internet_df = pd.json_normalize(df["internet"])
account_df = pd.json_normalize(df["account"])
```

```python
# Unir todo al DataFrame original
df = pd.concat(
    [df.drop(columns=["customer", "phone", "internet", "account"]),
     customer_df,
     phone_df,
     internet_df,
     account_df],
    axis=1
)
```

### Manejo de incosistencias

```python
#Convertir a numérico, poniendo NaN
df["Charges.Total"] = pd.to_numeric(df["Charges.Total"], errors="coerce")
```

```python
#Verificar cuantos valores NaN hay
df[df["Charges.Total"].isnull()]
```

```python
#Número de valores NaN
df["Charges.Total"].isnull().sum()
```

```python
#Cambiar los valores nulos por cero
df["Charges.Total"] = df["Charges.Total"].fillna(0)
```

```python
#Verificar nuevamento el número de valores NaN
df["Charges.Total"].isnull().sum()
```

```python
# Verificar el valor de NaN en la columna "Churn"
df["Churn"].isnull().sum()
```

```python
#No son NaN, por lo tanto se verifica si son strings vacíos:
(df["Churn"] == "").sum()
```

```python
#Como son valores vacíos, no se sabe si el cliente se fue o se quedó, por lo tanto para no intentar predecir, se eliminaran esas filas para no tener información falsa
df = df[df["Churn"] != ""]
```

```python
#Verificar que ya no haya valores vacíos
(df["Churn"] == "").sum()
```

```python
df["Churn"].value_counts()
```

### Creación de la columna cuentas diarias

```python
#Para generar el valor por día de las cuentas se dividrá por 30 días que es el promedio de días en un mes
df["Charges.Daily"] = df["Charges.Monthly"] / 30
df["Charges.Daily"] = df["Charges.Daily"].round(2)
```

```python
#Verificar los resultados:
df[["Charges.Monthly", "Charges.Daily"]].head()
```

### Estandarización y transformación de datos

```python
# Renombrar las columnas para que sea más fácil de entender el DataFrame
columnas_nuevas = {
    'gender': 'Género',
    'SeniorCitizen': 'Adulto mayor',
    'Partner': 'Tiene pareja',
    'Dependents': 'Dependientes',
    'tenure': 'Meses activo',
    'PhoneService': 'Servicio telefonico',
    'MultipleLines': 'Lineas múltiples',
    'InternetService': 'Servicio de internet',
    'OnlineSecurity': 'Seguridad online',
    'OnlineBackup': 'Respaldo online',
    'DeviceProtection': 'Protección dispositivo',
    'TechSupport': 'Soporte técnico',
    'StreamingTV': 'Televisión por cable',
    'StreamingMovies': 'Streaming películas',
    'Contract': 'Tipo contrato',
    'PaperlessBilling': 'Factura digital',
    'PaymentMethod': 'Forma de pago',
    'Charges.Monthly': 'Cargo mensual',
    'Charges.Total': 'Cargo total',
    'Charges.Daily': 'Cargo diario',
    'Churn': 'Evasión'
}

# Aplicar el renombramiento
df = df.rename(columns=columnas_nuevas)
df.head()
```

```python
#Identificar columnas con 2 valores
df.select_dtypes(include="object").nunique()
```

```python
#Convertir valores textuales Yes/No en valores binarios
columnas_binarias = df.columns[df.nunique() == 2]

for col in columnas_binarias:
    if set(df[col].unique()) == {"Yes", "No"}:
        df[col] = df[col].map({"Yes": 1, "No": 0})
```

```python
#Verificar las columnas binarias
df.head()
```
